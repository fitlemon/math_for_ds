ЗАЧЕМ СПЕЦИАЛИСТУ В DATA SCIENCE УЧИТЬ МАТЕМАТИКУ?

Зачем дата-сайентисту тратить время на изучение дисциплин высшей математики, если вместо этого можно просто импортировать библиотеку в Python, например знакомую нам sklearn, и построить свою модель без особых усилий?

Приведём несколько аргументов:

1. «Математика — царица наук»
   Математика — фундаментальная наука, на которой основаны практически все прикладные науки: от аэродинамики, благодаря которой летают самолеты, до искусственного интеллекта, включая машинное обучение. Именно математические законы лежат в основе большинства благ, которыми пользуется цивилизация на сегодняшний момент: от смартфонов и микроволновок до умных колонок и беспилотных автомобилей. 

   Изучение основ высшей математики позволит нам структурировать мышление. Мыслить в терминах матриц, функций и уравнений — это прямой путь к осмыслению как Machine Learning, так и Data Science в целом.

2. Математика позволит вам понимать, что вы делаете, и значительно повысить качество вашей работы.
   Чтобы стать профессионалом в любой области, нужно понимать принципы работы инструментов, которые вы используете. Так и с профессией Data Scientist. Любая модель машинного обучения — это в первую очередь математическая модель, основанная либо на теоремах, либо на математических эвристиках. 

   Линейная алгебра поможет вам понять, как работают линейные модели и нейронные сети. На теоремах теории вероятностей построены байесовские алгоритмы, деревья решений и многие другие алгоритмы машинного обучения. Математический анализ и вытекающая из него теория оптимизации — это фундамент, на котором основана теория построения моделей машинного обучения и их обучения.

   Знание того, как модель работает изнутри, позволит нам выбирать лучшие гиперпараметры, правильно интерпретировать результаты моделирования и разрабатывать лучшие модели.

   Благодаря пониманию статистических свойств данных вы сможете искать закономерности в них, делать выводы и правильно выбирать модель для решения поставленной задачи.

3. Математика — это язык, который понятен всем.
   На языке математики говорит всё мировое сообщество DS. Понимание терминологии и основ математики позволит вам понимать статьи, написанные на любом другом языке, практически не прибегая к помощи переводчика.

4. Математика позволит вам создавать собственные модели.
   Погрузившись в математический анализ и поняв принципы работы алгоритмов, вы, возможно, сможете построить собственный алгоритм, который превзойдёт по качеству, скорости работы или каким-то другим параметрам существующие. Может быть, вы войдёте в историю машинного обучения наряду с основоположником линейной регрессии Карлом Гауссом или создателем теории градиентного бустинга Джеромом Фридманом.

Наш путь в прекрасный мир высшей математики мы начинаем с раздела под названием линейная алгебра.

ЛИНЕЙНАЯ АЛГЕБРА В DATA SCIENCE

Линейная алгебра — это раздел алгебры, который изучает объекты линейной природы, векторы и векторные пространства, матрицы и линейные операторы, которые они порождают, и системы линейных уравнений.

Линейная алгебра лежит в основе решения всех основных задач Data Science: создания и настройки моделей, создания и обучения нейросетей, корреляционного анализа и т. д. В конце концов, даже табличный способ представления данных — это тоже фундаментальная основа линейной алгебры, матрица.

ЛИНЕЙНЫЕ МОДЕЛИ

Все линейные модели, такие как линейная и логистическая регрессия, берут своё начало из линейной алгебры. Сюда же относят метод опорных векторов (SVM).

МЕТОДЫ ПОНИЖЕНИЯ РАЗМЕРНОСТИ

В основе метода главных компонент (PCA), который мы неоднократно использовали для понижения размерности пространства признаков, лежит теория векторных пространств. Сингулярное разложение (SVD), о котором мы поговорим в модуле по рекомендательным системам, — это способ разложения матрицы на компоненты.

КОРРЕЛЯЦИОННАЯ МАТРИЦА

Напомним, что корреляция — это мера линейной зависимости между признаками. Если признаков несколько, то корреляция превращается в знакомую нам корреляционную матрицу. Матрицы и операции с ними изучает именно линейная алгебра.

ОБРАБОТКА ИЗОБРАЖЕНИЙ

Известно, что современные компьютеры работают только с числами 0 и 1. Тогда как в компьютере можно сохранить изображение, состоящее из различных цветов? Легко! Для этого изображение представляется в виде матрицы, отражающей интенсивность пикселей. Затем эта матрица может быть обработана для определения цветов, предметов и т. д.

Любая операция, которую вы хотите выполнить с этим изображением, вероятно, будет «под капотом» использовать линейную алгебру и матрицы.

Это лишь несколько очевидных примеров того, где линейная алгебра применяется в Data Science. На самом деле их намного больше, и мы поймём это, когда будем изучать, как методы машинного обучения устроены внутри.

ОСНОВЫ МАТЕМАТИЧЕСКОГО ЯЗЫКА: КВАНТОРЫ. ЧАСТЬ I

Прежде чем переходить к основной части материала, стоит вспомнить, что у математики есть свой «язык». Если вы читали статьи по математике, аналитике или машинному обучению, то могли заметить, что методы и функциональные зависимости там описываются на языке формул, который очень важно уметь понимать.

Возможно, со многими обозначениями вы уже знакомы, но стоит освежить их в памяти, чтобы дальше можно было без проблем понимать написанные формулировки. Разумеется, мы будем разбирать каждую тему на простых примерах, однако рядом всегда будем приводить и формальные обозначения. Для развития в Data Science (да и в любой науке, связанной с информатикой и математикой) большим бонусом является умение понимать математический язык, так как это даёт возможность читать различные статьи и понимать математические алгоритмы без дополнительных разъяснений.

Для начала познакомимся с кванторами и символами, позволяющими записывать математические формулировки:

- Символ читается как «для любого» и называется квантором всеобщности. Он заменяет собой слова «любой», «всякий», «для любого» и т. п.
- Символ читается как «существует» и называется квантором существования, а читается как «существует единственный» и называется квантором существования и единственности. заменяет слова «существует», «найдётся» и т. п.
- Знак импликации (или ) воспринимается как замена слов «следует», «влечёт», «вытекает», «имеет место», «выполняется».
- Знак равносильности или эквивалентности (или ) заменяет словосочетания «тогда и только тогда, когда», «в том и только том случае, когда», «если и только если».
- Двоеточие в математических формулах читается как «такой, что».
- Вертикальная черта означает «при условии, что».
